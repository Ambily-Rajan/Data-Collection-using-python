```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Referer': 'https://www.ambitionbox.com/'
}

webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_='companyCardWrapper')
```

```python
Name = []
Rating = []
Description = []
Highly_Rated_For = []
Critically_Rated_For = []

for i in company:
    name_element = i.find('h2')
    rating_element = i.find('span', class_='companyCardWrapper__companyRatingValue')
    description_element = i.find('span', class_='companyCardWrapper__interLinking')
    Highly_Rated_For_element = i.find('span', class_='companyCardWrapper__ratingHeader--high')
    Highly_Rated_For_value = i.find('span', class_='companyCardWrapper__ratingValues')
    Critically_Rated_For_element = i.find('span', class_='companyCardWrapper__ratingHeader--critical')
    Critically_Rated_For_value = i.find('span', class_='companyCardWrapper__ratingValues')

  
   
    # Check if the elements are found before accessing their text content
    
    if name_element:
        Name.append(name_element.text.strip())
    else:
        Name.append("N/A")

    if rating_element:
        Rating.append(rating_element.text.strip())
    else:
        Rating.append("N/A")

    if description_element:
        Description.append(description_element.text.strip())
    else:
        Description.append("N/A")
        
    if Highly_Rated_For_element and Highly_Rated_For_value:
        Highly_Rated_For.append( Highly_Rated_For_value.text.strip())
    else:
        Highly_Rated_For.append("N/A")
    
    if Critically_Rated_For_element and Critically_Rated_For_value:
        Critically_Rated_For.append( Critically_Rated_For_value.text.strip())
    else:
        Critically_Rated_For.append("N/A")
    
Expertise = []
Number_of_Employees = []
Type = []
Age = []
Location = []

# Extract information from the 'Description' list
for desc in Description:
    parts = desc.split(' | ')
    Expertise.append(parts[0])
    Number_of_Employees.append(parts[1])
    Type.append(parts[2])
    Age.append(parts[3])
    Location.append(parts[-1])
  
 # Create a DataFrame
df = pd.DataFrame({
    'Name': Name,
    'Rating': Rating,
    'Description': Description,
    'Expertise': Expertise,
    'Number_of_Employees': Number_of_Employees,
    'Type': Type,
    'Age': Age,
    'Location': Location,
    'Highly_Rated_For':Highly_Rated_For,
    'Critically_Rated_For':Critically_Rated_For
    
})

    
```

```python
all_data=pd.DataFrame()

for pages in range(1,503):
    
    url='https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(pages)
    
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Referer': 'https://www.ambitionbox.com/'
              }

    webpage=requests.get(url,headers=headers).text
    
    soup=BeautifulSoup(webpage,'lxml')
    
    company=soup.find_all('div',class_='companyCardWrapper')
    
    Name = []
    Rating = []
    Description = []
    Highly_Rated_For = []
    Critically_Rated_For = []

    for i in company:
        
        name_element = i.find('h2')
        rating_element = i.find('span', class_='companyCardWrapper__companyRatingValue')
        description_element = i.find('span', class_='companyCardWrapper__interLinking')
        Highly_Rated_For_element = i.find('span', class_='companyCardWrapper__ratingHeader--high')
        Highly_Rated_For_value = i.find('span', class_='companyCardWrapper__ratingValues')
        Critically_Rated_For_element = i.find('span', class_='companyCardWrapper__ratingHeader--critical')
        Critically_Rated_For_value = i.find('span', class_='companyCardWrapper__ratingValues')

  
   
    # Check if the elements are found before accessing their text content
    
        if name_element:
            Name.append(name_element.text.strip())
        else:
            Name.append("N/A")

        if rating_element:
             Rating.append(rating_element.text.strip())
        else:
             Rating.append("N/A")

        if description_element:
             Description.append(description_element.text.strip())
        else:
            Description.append("N/A")
        
        if Highly_Rated_For_element and Highly_Rated_For_value:
            Highly_Rated_For.append( Highly_Rated_For_value.text.strip())
        else:
            Highly_Rated_For.append("N/A")
    
        if Critically_Rated_For_element and Critically_Rated_For_value:
            Critically_Rated_For.append( Critically_Rated_For_value.text.strip())
        else:
            Critically_Rated_For.append("N/A")
            
            
    Expertise = []
    Number_of_Employees = []
    Type = []
    Age = []
    Location = []

    # Extract information from the 'Description' list
    for desc in Description:
        parts = desc.split(' | ')
        
        # Check if there are enough elements in the list
        if len(parts) >= 4:
            Expertise.append(parts[0])
            Number_of_Employees.append(parts[1])
            Type.append(parts[2])
            Age.append(parts[3])
            Location.append(parts[-1])
        else:
            # Handle the case where there are not enough elements
            Expertise.append("N/A")
            Number_of_Employees.append("N/A")
            Type.append("N/A")
            Age.append("N/A")
            Location.append("N/A")
        
            
  
     # Create a DataFrame
    df = pd.DataFrame({
        'Name': Name,
        'Rating': Rating,
        'Description': Description,
        'Expertise': Expertise,
        'Number_of_Employees': Number_of_Employees,
        'Type': Type,
        'Age': Age,
        'Location': Location,
        'Highly_Rated_For':Highly_Rated_For,
        'Critically_Rated_For':Critically_Rated_For
    
    })
    
    
    all_data=pd.concat([all_data,df],ignore_index=True)  
    
```

```python
all_data.drop('Description',axis=1,inplace=True)
```
