```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Referer': 'https://www.ambitionbox.com/'
}

webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_='companyCardWrapper')
```

```python
Name = []
Rating = []
Description = []
Highly_Rated_For = []
Critically_Rated_For = []

for i in company:
    name_element = i.find('h2')
    rating_element = i.find('span', class_='companyCardWrapper__companyRatingValue')
    description_element = i.find('span', class_='companyCardWrapper__interLinking')
    
     # Find Highly Rated For section
    highly_rated_for_section = i.find('span', class_='companyCardWrapper__ratingHeader--high')
    highly_rated_for_value = None
    if highly_rated_for_section:
        highly_rated_for_value = highly_rated_for_section.find_next('span', class_='companyCardWrapper__ratingValues')

    # Find Critically Rated For section
    critically_rated_for_section = i.find('span', class_='companyCardWrapper__ratingHeader--critical')
    critically_rated_for_value = None
    if critically_rated_for_section:
        critically_rated_for_value = critically_rated_for_section.find_next('span', class_='companyCardWrapper__ratingValues')


  
   
    # Check if the elements are found before accessing their text content
    
    if name_element:
        Name.append(name_element.text.strip())
    else:
        Name.append("N/A")

    if rating_element:
        Rating.append(rating_element.text.strip())
    else:
        Rating.append("N/A")

    if description_element:
        Description.append(description_element.text.strip())
    else:
        Description.append("N/A")
    
    Highly_Rated_For.append(highly_rated_for_value.text.strip() if highly_rated_for_value else "N/A")
    Critically_Rated_For.append(critically_rated_for_value.text.strip() if critically_rated_for_value else "N/A")
    


  
 # Create a DataFrame
df = pd.DataFrame({
    'Name': Name,
    'Rating': Rating,
    'Description': Description,
   
    'Highly_Rated_For':Highly_Rated_For,
    'Critically_Rated_For':Critically_Rated_For
    
})

    
```

```python
all_data=pd.DataFrame()

for pages in range(1,504):
    
    url='https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(pages)
    
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Referer': 'https://www.ambitionbox.com/'
              }

    webpage=requests.get(url,headers=headers).text
    
    soup=BeautifulSoup(webpage,'lxml')
    
    company=soup.find_all('div',class_='companyCardWrapper')
    
    Name = []
    Rating = []
    Description = []
    Highly_Rated_For = []
    Critically_Rated_For = []

    for i in company:
        
        name_element = i.find('h2')
        rating_element = i.find('span', class_='companyCardWrapper__companyRatingValue')
        description_element = i.find('span', class_='companyCardWrapper__interLinking')
        
           # Find Highly Rated For section
        highly_rated_for_section = i.find('span', class_='companyCardWrapper__ratingHeader--high')
        highly_rated_for_value = None
        if highly_rated_for_section:
                   highly_rated_for_value = highly_rated_for_section.find_next('span', class_='companyCardWrapper__ratingValues')
            

    # Find Critically Rated For section
        critically_rated_for_section = i.find('span', class_='companyCardWrapper__ratingHeader--critical')
        critically_rated_for_value = None
        if critically_rated_for_section:
                   critically_rated_for_value = critically_rated_for_section.find_next('span', class_='companyCardWrapper__ratingValues')

        
  
   
    # Check if the elements are found before accessing their text content
    
        if name_element:
            Name.append(name_element.text.strip())
        else:
            Name.append("N/A")

        if rating_element:
             Rating.append(rating_element.text.strip())
        else:
             Rating.append("N/A")

        if description_element:
             Description.append(description_element.text.strip())
        else:
            Description.append("N/A")
        
        Highly_Rated_For.append(highly_rated_for_value.text.strip() if highly_rated_for_value else "N/A")
        
        Critically_Rated_For.append(critically_rated_for_value.text.strip() if critically_rated_for_value else "N/A")
    
   
   
            
  
     # Create a DataFrame
    df = pd.DataFrame({
        'Name': Name,
        'Rating': Rating,
        'Description': Description,
        'Highly_Rated_For':Highly_Rated_For,
        'Critically_Rated_For':Critically_Rated_For
    
    })
    
    
    all_data=pd.concat([all_data,df],ignore_index=True)  
    
```
